### `Section 2` - Preparing the Data and Cleaning It...

- **Missing Values**

```python
# We are dropping the `Id` column because it serves as a unique identifier for each row, 
# And does not contribute any meaningful information for predicting house prices.
data = data.drop(columns=['Id'])
```

```python
'''
    - We replaced the null values in the below columns with meaningful placeholders.
    - Based on the data description file, null values in these columns indicate the absence of the respective feature.
    - This is the best way to deal with these features, as the data description file indicates that null values mean
      the feature is not present. 
    - Therefore, the best solution is to fill these null values with specific strings that explicitly indicate the absence
      of the feature, ensuring the dataset is consistent, clear, and that we don't lose any important information.
'''
data['PoolQC']       = data['PoolQC'].fillna('No Pool')
data['MiscFeature']  = data['MiscFeature'].fillna('None')
data['Alley']        = data['Alley'].fillna('No Alley Access')
data['Fence']        = data['Fence'].fillna('No Fence')
data['MasVnrType']   = data['MasVnrType'].fillna('None')
data['FireplaceQu']  = data['FireplaceQu'].fillna('No Fireplace')
data['GarageType']   = data['GarageType'].fillna('No Garage')
data['GarageFinish'] = data['GarageFinish'].fillna('No Garage')
data['GarageQual']   = data['GarageQual'].fillna('No Garage')
data['GarageCond']   = data['GarageCond'].fillna('No Garage')
data['BsmtQual']     = data['BsmtQual'].fillna('No Basement')
data['BsmtCond']     = data['BsmtCond'].fillna('No Basement')
data['BsmtExposure'] = data['BsmtExposure'].fillna('No Basement')
data['BsmtFinType1'] = data['BsmtFinType1'].fillna('No Basement')
data['BsmtFinType2'] = data['BsmtFinType2'].fillna('No Basement')
```

```python
'''
    - The `Electrical` column had only one null value out of the 1460 rows present in the dataset.
    - Given that this represents less than 1% of the data, it is suitable and logical to drop the row
      containing the null value.
    - Removing such a small portion of the dataset will not significantly impact the analysis or the results 
      of the machine learning models.
'''
data = data.dropna(subset=['Electrical'])
```

```python
'''
    - The `GarageYrBlt` column indicates that there is no garage present for the property.
    - By replacing null values with zero, we explicitly indicate the absence of a garage, 
      making the dataset consistent and easier to interpret.
    - This approach ensures that the data remains complete and that the absence of a garage
      is clearly represented in the analysis.
'''
data['GarageYrBlt'] = data['GarageYrBlt'].fillna(0)
```

```python
'''
    - We replaced the null values in the `MasVnrArea` column with the mode value.
    - This decision was made because most of the values are concentrated around the mode,
      making it a logical choice to maintain the distribution and central tendency of the data.
'''
mode_value         = data['MasVnrArea'].mode().iloc[0]
data['MasVnrArea'] = data['MasVnrArea'].fillna(mode_value)

sns.histplot(data['MasVnrArea'])

plt.show()
```

```python
'''
    - Replacing null values in the `LotFrontage` column with the median value.
    - The median is a better measure of central tendency for `LotFrontage` as it is
      less affected by outliers compared to the mean.
    - This makes the median more suitable for filling in missing values, ensuring that
      the central tendency and distribution of `LotFrontage` remain 
      accurate and representative of the dataset.
'''
median_value        = data['LotFrontage'].median()
data['LotFrontage'] = data['LotFrontage'].fillna(mode_value)

sns.kdeplot(data['LotFrontage'], fill=True)

plt.show()
```
